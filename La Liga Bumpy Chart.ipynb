```python
import pandas as pd
import matplotlib.pyplot as plt
import mplsoccer
import highlight_text
from mplsoccer import Bumpy
import numpy as np
import matplotlib.image as mpimg
```

```python
df = pd.read_csv('La Liga Standings.csv')
```

```python
image = mpimg.imread('La Liga Logo.png')
```

```python
df = df.T
```

```python
df.columns = df.iloc[0]
df = df.iloc[1:]
```

```python
df
```

```python
week = ['Week ' + str(num) for num in range (1,19)]
```

```python
week
```

```python
highlight_dict = {
    'Barcelona':'#008afd',
    'Atletico Madrid':'#cb3524',
    'Real Madrid':'white',
    'Athletic Club':'#edbb00'
}
```

```python
bumpy = Bumpy(
    scatter_color = 'white',
    line_color = 'white',
    rotate_xticks = 90,
    rotate_yticks = 0,
    scatter_primary = 'D',
    show_right = True,
    plot_labels = True,
    alignment_yvalue = .1,
    alignment_xvalue = .065
)
```

```python
fig, ax = bumpy.plot(
    x_list = week,
    y_list = np.linspace(1,20,20).astype(int),
    values = df,
    secondary_alpha = .5,
    highlight_dict = highlight_dict,
    figsize = (18,18),
    ylim = (-.1,23),
    lw = 2.5
)

fig.text(s = 'La Liga Top 4 Positions until MatchDay 18.', x = .13, y = .84,
         c = 'white', size = 24, weight = 'bold')

highlight_text.fig_text(x = .13, y = .82,
                        s = 'Comparing <Barcelona>, <Atletico Madrid>, <Real Madrid> and <Athletic Club>.',
                        highlight_textprops = [
                            {"color" : '#008afd'},
                            {"color" : '#cb3524'},
                            {"color" : 'white'},
                            {"color" : '#edbb00'}
                        ],
                        fontsize = 20,
                        color = 'white'
                       )

#ax.imshow(image, extent=[.13,0.5, .84,0.9], aspect = 1, alpha=1)

endnote = "- Rishav Dey"
plt.figtext(0.9, 0.05, endnote, ha="center", va="top", fontsize=15, color="white")
```
