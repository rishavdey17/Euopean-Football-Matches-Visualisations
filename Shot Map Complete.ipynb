```python
data_latest = pd.read_csv('Mallorca vs Barcelona.csv')
```

```python
data_latest
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch, VerticalPitch
```

```python
goals_barca = data_latest[data_latest["Result"] == 'Goal'].copy()
non_goals_barca = data_latest[data_latest["Result"] != 'Goal'].copy()
```

```python
goals_barca = data_latest[data_latest["Result"] == 'Goal'].copy()
missed_barca = data_latest[data_latest["Result"] == 'Missed'].copy()
blocked_barca = data_latest[data_latest["Result"] == 'Blocked'].copy()
woodwork_barca = data_latest[data_latest["Result"] == 'Woodwork'].copy()
saved_barca = data_latest[data_latest["Result"] == 'Saved'].copy()
```

```python
pitch = VerticalPitch(pitch_type='custom', pitch_width = 100, pitch_length = 100,
                      pitch_color='#0a111a', line_color='#c7d5cc', half = True)
fig, ax = pitch.draw(figsize=(13,8.5), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#0a141a')

#pitch.draw(ax=ax)
#plt.gca().invert_yaxis()

goals = pitch.scatter(goals_barca['x'],
                      goals_barca['y'],
                      s = goals_barca['xG'] * 500 + 100,
                      c ='g',
                      edgecolor = 'white',
                      ax=ax)

missed = pitch.scatter(missed_barca['x'],
                       missed_barca['y'],
                       s = missed_barca['xG'] * 500 + 100,
                       c ='r',
                       edgecolor = 'white',
                       ax=ax)

blocked = pitch.scatter(blocked_barca['x'],
                        blocked_barca['y'],
                        s = blocked_barca['xG'] * 500 + 100,
                        c ='r',
                        edgecolor = 'white',
                        ax=ax)

woodwork = pitch.scatter(woodwork_barca['x'],
                         woodwork_barca['y'],
                         s = woodwork_barca['xG'] * 1000 + 100,
                         c ='w',
                         edgecolor = 'white',
                         ax=ax)

saved = pitch.scatter(saved_barca['x'],
                      saved_barca['y'],
                      s = saved_barca['xG'] * 500 + 100,
                      c ='y',
                      edgecolor = 'white',
                      ax=ax)


plt.title("Barcelona's Shots vs Mallorca.", fontsize = 24, c = 'w')

plt.scatter(x = 98, y = 102, s = 200, c = 'g')
plt.text(x = 96, y = 101, s = f'Goal', c = 'g', fontsize = 15, fontweight = 'bold')

plt.scatter(x = 77, y = 102, s = 200, c = 'w',)
plt.text(x = 75, y = 101, s = f'Woodwork', c = 'w', fontsize = 15, fontweight = 'bold')

plt.scatter(x = 48, y = 102, s = 200, c = 'y',)
plt.text(x = 46, y = 101, s = f'Saved', c = 'y', fontsize = 15, fontweight = 'bold')

plt.scatter(x = 22, y = 102, s = 200, c = 'r',)
plt.text(x = 20, y = 101, s = f'Missed/Blocked', c = 'r', fontsize = 15, fontweight = 'bold')

plt.text(x=100, y=40, s ='Shots', fontsize = 15, fontweight = 'bold', c = 'w', ha = 'left')
plt.text(x=97.5, y=35, s = '20', fontsize = 15, fontweight = 'bold', c = 'w', ha = 'left')

plt.text(x=80, y=40, s ='Expected Goals (xG)', fontsize = 15, fontweight = 'bold', c = 'w', ha = 'left')
plt.text(x=70, y=35, s = '5.38', fontsize = 15, fontweight = 'bold', c = 'w', ha = 'left')

plt.text(x=40, y=40, s ='Goals', fontsize = 15, fontweight = 'bold', c = 'w', ha = 'left')
plt.text(x=37.5, y=35, s = '5', fontsize = 15, fontweight = 'bold', c = 'w', ha = 'left')

plt.text(x=15, y=40, s ='xG per Shot', fontsize = 15, fontweight = 'bold', c = 'w', ha = 'left')
plt.text(x=10, y=35, s = '0.269', fontsize = 15, fontweight = 'bold', c = 'w', ha = 'left')
```
